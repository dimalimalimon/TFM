```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
loss=[]
with open("./log_train.txt", "r") as f:
    lines=f.readlines()
for l in lines:
    l=l.strip().split(" ")
    if "Test" not in l and len(l)!=1:
        loss.append(float(l[-1]))
```

```python
plt.plot(np.arange(len(loss)),loss)
plt.title("Total loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
```

```python
plt.plot(np.arange(100,len(loss)),loss[100:])
plt.title("Epoch 100+ loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
```

```python

def get_loss(subtype):
    loss=[]
    with open("./log_"+subtype+".txt", "r") as f:
        lines=f.readlines()
    for l in lines:
        l=l.strip().split(" ")
        if "Test" not in l and len(l)!=1 and "loss:" in l:
            loss.append(float(l[-1]))
    return(loss)
loss_g=get_loss("G")
loss_a=get_loss("A")
loss_e=get_loss("E")
```

```python
fig, axs = plt.subplots(1, 3, figsize=(12, 4))

# Plot data on each subplot
axs[0].plot(np.arange(10,len(loss_g)), loss_g[10:])
axs[0].set_title('Loss G')
axs[0].set_title('Loss G')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Loss')

axs[1].plot(np.arange(10,len(loss_a)), loss_a[10:])
axs[1].set_title('Loss A')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Loss')

axs[2].plot(np.arange(10,len(loss_e)), loss_e[10:])
axs[2].set_title('Loss E')
axs[2].set_xlabel('Epoch')
axs[2].set_ylabel('Loss')
# Adjust spacing between subplots
plt.tight_layout()
```
